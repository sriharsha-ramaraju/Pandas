```python
#we can 1)either drop the column or the rows with null values or 2)fill the null (missing) values using mean, median

# Filling missing values using fillna(), replace() and interpolate()

df_fillna1=df.fillna(0) #fills all the missing values by 0s
df_fillna1.isnull().sum()#shows no missing value

df_fillna2=df.fillna(method='pad') #replaces the na value with the value before current na
df_fillna2.isnull().sum()#shows no missing value

df_fillna2=df.fillna(method='bfill') #replaces the na value with the value after current na
df_fillna2.isnull().sum()#shows no missing value
```

 Filling Missing Values

```python
#let's do this on some sample data
# dictionary of lists 
dict = {'First Score':[100, 90, np.nan, 95], 
        'Second Score': [30, 45, 56, np.nan], 
        'Third Score':[np.nan, 40, 80, 98]} 
  
# creating a dataframe from dictionary 
df = pd.DataFrame(dict) 
  
# filling  null value using fillna() function   
print('------------- forwardfill')
print(df.fillna(method ='ffill'))#replaces the na value with the value before current na
print('------------- backwardfill')
print(df.fillna(method ='bfill'))#replaces the na value with the value after current na

################# using replace
print('------------- replace')
#replacing with predetermined value
print(df.replace(to_replace = np.nan, value = -1000) )

print('-------------using drop.na-removing obs with na')#observasation with na values are dropped

print(df.dropna())

print('-------------using drop.na-removing columns with na')#observasation with na values are dropped

print(df.dropna(axis=1,thresh=2)) #every column has at least 2 na values..so all columns are removed

```

```python
dict = {'First Score':[100, 90, np.nan, 95], 
        'Second Score': [30, 45, 56, np.nan], 
        'Third Score':[np.nan, 40, 80, 98]} 
  
# creating a dataframe from dictionary 
df = pd.DataFrame(dict) 

df.fillna({'First Score': 1000,
          'Second Score': 3000,
          'Third Score': 8000}
)#we can replace each column separately
```
